# <span style="color: #87BBA2">===   Git e GitHub: compartilhando e colaborando em projetos   ===</span>

## <span style="color: #87BBA2">01. Compartilhando projetos</span>

### SOBRE GIT INIT
O comando git init deve ser utilizado para converter um diretório existente no computador, que geralmente é o diretório de algum projeto, em um repositório Git. No entanto, seu uso requer atenção e cuidado para evitar problemas indesejados. Vamos entender melhor o que o comando git init faz.

O que é o comando git init?
Para que possamos realizar o controle de versão de um projeto, registrando as mudanças realizadas nele ao longo do tempo, devemos, primeiramente, transformar o diretório do projeto em um repositório Git. O comando git init é utilizado para esse objetivo, devendo ser executado apenas uma vez. Quando executado, ele configura o diretório atual para ser rastreado pelo Git, inicializando um repositório vazio.

Executando o comando git init
Suponha que no seu computador exista um diretório chamado meu-projeto, que represente um projeto pessoal seu e que você deseja transformar em um repositório do Git. Para isso, você pode abrir esse diretório no VSCode, abrir uma janela do terminal e executar o comando git init. A saída será algo como:

![alt text](https://cdn1.gnarususercontent.com.br/1/4341/ea145a08-0114-4a5f-bb96-537fff10d7b5.png)

Terminal do VSCode com o comando git init sendo executado no diretório "~/Projetos/alura/meu-projeto" e recebendo como saída a mensagem "Initialized empty Git repository in /home/rodrigo/Projetos/alura/meu-projeto/.git/"
Repare na imagem anterior que o comando git init foi executado no terminal do VSCode. Observe também que no terminal é indicado qual o diretório no qual o comando foi executado, que no exemplo foi em: ~/Projetos/alura/meu-projeto. É importante se atentar a isso, pois o comando git init transforma o diretório atual em um repositório do Git, logo ele deve ser executado dentro do diretório do projeto e não em outros diretórios do computador.

Ao executar o comando, note que a saída no terminal foi a mensagem Initialized empty Git repository in /home/rodrigo/Projetos/alura/meu-projeto/.git/. Essa mensagem indica que o comando foi executado corretamente e um repositório local do Git foi criado com sucesso nesse diretório. A partir desse ponto, já podemos trabalhar no projeto, adicionando arquivos, realizando modificações e registrando as mudanças no Git.

Cuidados com o comando git init
Aprendemos que o comando git init serve para criar um novo repositório Git e por isso deve ser executado apenas uma única vez. Ou seja, se um diretório já for um repositório Git, não faz sentido rodar novamente o comando git init. Esse é um erro bastante comum de ser cometido.

Se você executar o comando git init em um diretório que já foi inicializado como um repositório Git, a seguinte mensagem será exibida:

`Reinitialized existing Git repository in /home/rodrigo/Projetos/alura/meu-projeto/.git/`


Isso indica que o Git reinicializou um repositório já existente, ou seja, o comando git init foi executado em um diretório que já era um repositório Git.

Caso você tenha cometido esse equívoco, não precisa se preocupar, pois todo o histórico de mudanças e commits no projeto não será apagado. O Git detecta que o diretório já era um repositório Git e com isso o comando não tem efeito nenhum.

Na dúvida, antes de executar o comando git init, execute primeiramente o comando git status. Se aparecer a mensagem fatal: not a git repository (or any of the parent directories): .git, isso significa que o diretório atual não é um repositório Git e você pode então executar o comando git init.

### SINCRONIZANDO REPOSITÓRIOS

Dado o `git init`, adicionamos os arquivos com o `git add` (lembrando, podemos adicionar unitariamente ou adicionar por completo com `git add .`). Alguns alertas de compatibilização entre Windows e Linux, mas, nada prejudicial.

Em seguida, quando daremos o `git commit -m "mensagem"`, receberemos um aviso de erro representando a não identificação do autor do commit, ou seja, quando isso acontece significa que não configuramos as credenciais de nosso git, retornando na própria linha de código comandos para realizar esse credenciamento.

#### CREDENCIAMENTO

Utilizando os comandos recomendados pelo git no terminal, executamos:
```JAVASCRIPT
git config --global user.mail "e-mail_usado_no_github@email.com"
git config --global user.name "Seu Nome"
```

Utilizando o parametro `--global` realizamos a configuração globalmente, ou seja, qualquer repositório que criamos neste mesmo computador utilizará dessas credenciais cadastradas.

Feito isso, é possivel dar prosseguimento às configurações do repositório remoto, como montar o `git branch -M main`.

#### SINCRONIZANDO
Agora, para realizarmos a sincronização, nós adicionamos um ponto remoto através do comando:
```JAVASCRIPT
git remote add origin link_https_ou_ssh
```
Idealmente utiliza-se o SSH ao invés de HTTPS, no geral, é mais recomendado para fazer esse link com o GitHub.

Ainda não conectamos com o repositório remoto, mas sim adicionamos um ponto para conexão.

Utilizando agora o comando para pushar (ou, empurrar) o conteúdo que está sendo rastreado pelo git localmete, utilizamos o comando `git push`. Neste momento que o programa provavelmente tentará realizar a conexão.

Precisamos também, definir a origem de onde vamos empurrar o conteúdo:
```JAVASCRIPT
git push -u origin main
```
`-u` significa `--set-upstream`, ou seja, configurar o caminho da branch atual para a branch remota. Estamos dizendo:
- Git, empurre as alteração confirmadas (commitadas) para o caminho informado e configure este caminho para ser na ramificação main da origem.
  - Onde, nós adicionamos a origem com o link_https_ou_ssh informado no `git remote add origin link_https_ou_ssh`.

Com isso, configuramos, então para onde estamos empurrando nossos documentos confirmados e os comandos (ao menos para essa branch) não precisarão mais ter seu caminho configurado (`--set-upstream`).

#### AUTENTICANDO

Quando usamos o protocolo SSH e sendo a primeira vez que estamos utilizando o git em uma máquina, aparecerá uma mensagem de erro logo em seguida do `push`, onde informa-se que que `This key is not known by any other names.`, ou seja, não configuramos nossa senhado GitHub em nosso repositório local, pois, agora o Git sabe quem somos nós mas o GitHub não sabe se a pessoa cadastrada é de fato a pessoa que detem a contam no Github.

Respondendo yes para a pergunta de se gostariamos de nos conectar mesmo assim, retorna-se a mensagem indicando que o github foi adicionado para a lista de hóspedes e que nossa permissão está negada. Isso significa que em nossa conta GitHub não há uma chave SSH cadastrada que corresponda a chave SSH que está tentando conectar ao repositório remoto com aquela conta. Ou seja, ele não deixará você utilizar aquela conta (e seus repositórios) se sua chave SSH do teu PC não estiver cadastrada ali.

##### Autorizando chave SSH no GitHub
Agora, vamos no menu superior direito e vamos em **settings > SSH and GPG Keys > Add new SSH Keys**

>- Title pode ser qualquer nome;
>- Key type: é uma Authenticaton key;
>- Key: aí entra nossa chave SSH.

##### Gerando uma chave SSH
[Documentação do GitHub explicando como gerar uma chave SSH](https://docs.github.com/en/authentication/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent)

Código para gerar a chave:
```JAVASCRIPT
ssh-keygen -t ed25519 -C "your_email@example.com"
```
Após rodar esse comando:
- Perguntará qual pasta deseja salvar o arquivo, entre parenteses está a pasta padrão e dando um enter salvaremos no caminho padrão. Se quisermos um caminho diferente é só informar neste momento.
- ENTER PASSPHRASE: perguntando se quer inserir uma senha. Caso você coloque uma senha, para diversas operações no GitHub será perguntada essa senha. Não há grandes problemas em deixá-la vazia (só dar espaço);
- PEDE PARA CONFIRMAR A SENHA: confirme, se colocou nenhuma senha só dar outro espaço.
- **CHAVE GERADA**: Uma pública (com extensão.pub) e outra privada. Usaremos a publica.
- As chaves, por padrão, estão no user/nome_usuario/.ssh, do caminho do seu computador.

##### Adicionando chave SSH no GitHub
Abrindo o arquivo `.pub` e salvando **TODO SEU CONTEUDO (incluindo e-mail)**, vamos no caminho do GitHub para adicionar chave SSH e colamos no campo de `Key:` e concluímos a adição.

Com isso, o GitHub ao consultar nossa chave SSH que está em nossa máquina conseguirá parear com o que está autorizado na conta, e assim, realizar as operações com maior segurança. Seguindo todas essas operações, o GitHub saberá que nessa máquina há uma chave SSH, executará a pareação com essa chave e partirá a confiar em nossa máquina.

## <span style="color: #87BBA2">02. Colaborando em projetos</span>

### CLONANDO UM REPOSITÓRIO

Utilizando o comando `git clone url_do_projeto` cria uma pasta com o projeto clonado juntamente com todo o histórico git do projeto. Clonando um repositório remoto você automaticamente, em seu repositório local, já possui seu ambiente git configurado e rastreando as mudanças feitas no repositório local frente ao estado do repositório remoto (ao menos, a atualização que foi clonada ou as atualizações buscadas com o `git pull`).

### REALIZANDO UM COMMIT

Para verificar o estado atual do repositório local frente a ultima atualização buscada do repositório remoto, usamos os `git status`.

Nele, vemos como está o estado atual do meu repositório git local em comparação com o repositório remoto, por exemplo, se temos alterações que fizemos que não subimos no repositório origem (remoto).

#### adicionando e commitando repositório clonado
Após visualizar o estado atual do repositório local, caso temos alterações a adicionar ou modificações a confirmar, podemos dar o `git add` da mesma maneira quando o repositório é nosso.

Adições feitas, podemos agora realizar o commit, também da mesma forma de quando o repositório remoto é nosso, mas, o `git push` (colaboração), demanda mais coisas.

#### git log: vendo todas as alterações feitas ao longo do projeto
Levantando relação completa das alterações realizadas
```JAVASCRIPT
git log
```

#### indicativos ao lado do nome dos arquivos

- M: A letra M representa o estado Modified, do português modificado. Isso significa que o arquivo já existia no repositório, mas que recebeu alguma modificação que ainda não foi registrada no Git.
- U: A letra U representa o estado Untracked, do português não rastreado. Isso significa que o arquivo ainda não existia no repositório e que ainda não teve seu registro (commit) feito no Git.
- A: Added, significa que o arquivo foi adicionado ao para ser commitado mas o commit ainda não foi realizado. Ou seja, está aguardando ser commitado.

### ENVIANDO COMMITS
Realizando o git clone, já importamos os parametros do git remoto, como qual é sua origem e afins, então, não precisamos relizar essa configuração novamente.

### git push origin main
Para podermos colaborar com um projeto, agora, precisamos dar o comando `git push` e indicar para onde:
```JAVASCRIPT
git push origin main // ou seja, empurre para o link adicionado como origin para a branch main
```
Porém, retornou-se um erro de permissão! Isso é normal, pois, apesar do repositório ser público, ele é público **apenas para leitura**, para escrita, só o criador do repositório e colaboradores podem realizar.
> O git faz isso pois quando alguém dá push, o git conseguiu ver quem é o usuário que está fazendo isso, pega suas credencias e bate para ver se é o adminstrador ou colaborador e, se não for nenhum destes, ele retorna o erro acima citado.

Com isso, o administrador do repositório adiciona a pessoa como colaboradora e agora, quando ela for dar `git push origin main`, os commits subirão ao repositório remoto sem problemas.

Commit feito pelo colaborador, o repositório local dos demais colaboradores não é atualizado automaticamente, agora veremos como poderemos realizar essa atualização.

### BAIXANDO NOVOS COMMITS

Utilizando o comando `git pull` conseguimos atualizar nosso repositório local com as alterações feitas por outras pessoas no repositório remoto:
```JAVASCRIPT
git pull origin main // mesma lógica de indicar a localização
```

#### Comandos comuns utilizados no dia a dia com Git
- git status|: verificar quais arquivos foram modificados;
- git add: adicionar as mudanças;
- git commit: registrar as mudanças no repositório;
- git push: subir as mudanças para o repositório remoto;
- git pull: baixar as mudanças para o repositório local.




## <span style="color: #87BBA2">03. Utilizando o Git na IDE</span>

### GIT NO VSCODE
Demonstrou-se a execução de tudo o que já fizemos anteriormente, mas, agora, no próprio VS CODE, na aba Source Control. Nela, executamos os mesmos fluxos que fizemos com os comandos, mas, visualmente.

Para adicionar arquivos é clicando no "+" a frente deles, para visualizar o arquivo, é clicando em sua pastinha, para commitar as modificações adicionadas, é clicando em "commit" e para escrever mensagem é na caixa de mensagem acima do commit.

Para dar o push, é clicando em Sync no mesmo local do botão do commit, após este for realizado.

### SIMULANDO UM CONFLITO
Para simular um conflito, foi feita uma alteração e em seguida, na aba do Git (Source Control), clicou-se nos 3 pontinhos e selecionou git pull. O git acusou conflito (ou seja, alguém subiu no repositório remoto alguma alteração que não estava em nosso repositório local anteriormente e a alteração que estamos tentando subir entra em conflito com o que está no repositório remoto).
- Ou seja, não é uma mudança, simplesmente, do que está no remoto frente ao que estava no nosso local, e sim, que há alguma alteração que foi feita na mesma linha de código que nós fizemos e que nós não estavamos acompanhando em nosso projeto.
- Quando isso acontece, o Git mostra o `diff` para nós pedindo para nós escolhermos o que fazer.

### RESOLVENDO CONFLITOS
Conflito elencado, podemos "resolver com o Merge Editor" ou manualmente. Na aula, vimos o manual:

#### RESOLVENDO CONFLITOS MANUALMENTE
Em verde, Git mostra a nossa alteração e em azul a alteração de outro desenvolvedor.

Agora, conversando com a equipe ou entre os responsáveis, decide-se qual será o código que permanecerá e qual sairá. No caso, decidiram por manter o código do outro desenvolvedor.

Para resolver o conflito, agora, precisamos apagar todos os códigos conflitantes recusados deixando só o que foi aceito, apagando também todas as marcações deixadas pelo Git, já que, quando há esse tipo de conflito o Git joga tudo no mesmo código.

Por fim, conflito resolvido, **necessitá-se realizar o `git add` e `git commit` dessa resolução e, em seguida, o `git push`**, pois uma resolução de um conflito é um novo processo de commit. Ou seja, subiremos além do commit da alteração que nós fizemos, subiremos o commit da resolução do conflito (que por sinal é resultado de um merge).

## <span style="color: #87BBA2">04. Voltando no tempo</span>

### REVERTENDO UM COMMIT

Caso queiramos reverter um commit, ou seja, retornar ao commit anterior ao que será revertido, usaremos o comando `git revert`
- este comando pede a **id do commit a ser revertido, isso é obtido através do `git log`**
```JAVASCRIPT
git revert id_do_commit_a_ser_revertido
```

`git revert` não apaga um commit, mas sim, analisa o commit a ser revertido, quais foram suas alterações, desfaz suas alterações e **cria um arquivo novo com essas mudanças feitas**, ou seja, um arquivo novo que reflete o estado anterior do commit a ser revertido.

Como criamos um arquivo novo, segue o processo de commitar e pushar.
- Não acho que precisará adicionar, pois estamos retornando a um estado já conhecido pelo Git.

### APAGANDO COMMIT
Para apagar o commit e seu registro, usamos o comando `git reset` com o parametro `--hard` e o id do commit **anterior** ao que queremos apagar. Ou seja, não é um delete, e sim **resetar a um estado**, ou seja, temos que indicar o id do commit que queremos resetar para este estado:
```JAVASCRIPT
git reset --hard id_do_commit_desejado
```

Com isso, apagamos o commit realizado. Durante o curso, **sempre estamos vendo o uso comum do `git log`**, principalmente para buscar o id dos commits.

`git reset` possui mais comportamentos a depender do seu parametro. Com o `--hard`, apagamos tudo: o commit e seu registro.

#### Boas praticas
Importante dizer: **Não é ideal realizarmos essa operação quando o commit foi subiu para o remoto, pois isso pode com altas chances gerar confusão**. Então, o ideal é fazermos isso em nossos projetos individuais ou em nosso repositório local de commits que não subiram para o remoto.

#### Documetanção sobre git reset
[Documentação oficial](https://git-scm.com/docs/git-reset/pt_BR) explicando mais funcionalidades do `git reset` e seu comportamento com diferentes parametros.

### ALTERANDO O ULTIMO COMMIT
Utilizando o comando `git commit` com o parametro `--amend`, conseguimos corrigir **o ultimo commit** feito, seja em sua mensagem, utilizando `git commit --amend -m "mensagem corrigindo a anterior`, ou adicionando mais arquivos com o `git add .`, `git commit -amend`.

Mas importante: Alterações no histórico de commits como o reset ou amend deve ser feito **com bastante cautela**. Recomenda-se fazer essas alterações apenas quando muito necessárias e **quando o commit ainda não tornou-se publico**, pois o histórico de commits tem de ser um retrato preciso do andamento de um projeto. Executar alterações quando publicadas tem grandes chances de poluir o histórico de commits e/ou gerar conflitos.

> Em situações de colaboração em equipe, é essencial manter a integridade do histórico de commits, pois qualquer modificação em um commit que outras pessoas estejam trabalhando pode resultar em conflitos e dificuldades na colaboração.

> É recomendável evitar a modificação excessiva do histórico de commits, uma vez que isso pode tornar o histórico confuso. O histórico deve ser uma representação precisa do progresso do projeto ao longo do tempo.

### IGNORANDO ARQUIVOS NO REPOSITORIO

Para indicarmos ao git pastas e arquivos para serem ignorados, ou seja, para o git não rastrear, antes de relizarmos o o `git add .`, criamos um arquivo chamado `.gitignore` e dentro dele passamos as instruções de quais pastas ou arquivos ele não deverá rastrear em nenhuma hipotese.

Algo interessante é que cada tecnologia gera arquivos padrão que, geralmente, não gostariamos de rastrea-los ou envia-los para o repositório remoto, como arquivos de log, configuração temporária ou sigilosos.

Para isso, podemos utilizar o [gitignore.io da Toptal](https://www.toptal.com/developers/gitignore), que gera gitignores padrão para cada tecnologia.

### COMPARTILHANDO CÓDIGOS COM GIST
No mesmo local onde podemos adicionar um novo repositório (lá no +), podemos publicar também um novo `gist`, ou seja, arquivos especificos de código.
- É uma forma mais resumida para publicarmos, onde não envolve o controle robusto de versionamento que um repósitório tem, é algo bem mais conciso e pontual mesmo, onde, quando formos publica (ou deixá-los secretos), somente é demonstrando os arquivos e os códigos mesmo.
- Sobre controle de versão, tem-se uma forma mais simplificada que seriam suas "revisões". Possui uma interface parecida com o dos repositórios em termos de adição e exclusão de linhas de código, mas lembre-se, tudo mais conciso.
- Basicamente, gist é uma versão pocket do repositório.
