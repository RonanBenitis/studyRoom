#### Indicador para não aceitar nulo
Quando atribuimos uma escolha com o `Console.ReadLine()`, uma tarja verde sob o comando apareceu indicando que, além de string, valores nulos estão sendo capturados por este comando. Ao adicionar `!` no fim do comando, dizemos que não queremos aceitar valores nulos.

```CS
string opcaoEscolhida = Console.ReadLine()!;
```

Note que para capturar algo digitado pelo usuário, usamos Console.ReadLine() neste caso, que capturará a linha inteira. Correspondencia Java: `String opcaoEscolhida = scanner.nextLine()`. 


#### Console Writing and Reading
- Console.WriteLine() = Printa uma informação e no fim pula uma linha. Correspondencia Java: `System.out.println()`
- Console.Write() = Printa uma informação e mantêm-se na mesma linha. Correspondencia Java: `System.out.print()`
- Console.ReadLine() = Captura uma linha digitada pelo usuário. Correspondencia Java: `Scanner scanner = new Scanner(System.in)` `scanner.nextLine()`

#### Casting explícito
Ao utilizar `Read.Line`, não podemos implicitamente castar entre tipos diferentes, por exemplo, capturá-lo como `int`. Para isso, precisamos explicitamente castá-lo com `Parse`, ou seja, `int.Parse(valor_a_ser_parseado)`

#### Switch/Case
O uso do Switch/Case é o mesmo que já vimos em outras linguagens

Veja que sua aplicação é mais interessante do que a opção de if/else para este caso.

Interessante ler o Switch/Case como um disjuntor, ativando um bloco de comandos CASO algo aconteça, ou seja, é um Disjuntor para exec

```CS
switch (opcaoEscolhidaNum)
{
    case 1:
        Console.WriteLine("Você escolheu a opção " + opcaoEscolhidaNum);
        break;
    case 2:
        Console.WriteLine("Você escolheu a opção " + opcaoEscolhidaNum);
        break;
    case 3:
        Console.WriteLine("Você escolheu a opção " + opcaoEscolhidaNum);
        break;
    case 4:
        Console.WriteLine("Você escolheu a opção " + opcaoEscolhidaNum);
        break;
    case -1:
        Console.WriteLine("Tchau tchau :)");
        break;
    default:
        Console.Write("Opração inválida");
        break;
}
```

### INTERPOLAÇÃO: $-STRING E FORMAT STRING

#### $-STRING
Assim como no Python, podemos abrir um micro ambiente C# usando `$"{micro_ambiente}"`, interpolando string com o valor das variaveis inseridas no campo do "micro_ambiente".

```CS
Console.WriteLine($"Olá, {nome}!");
```
> Neste caso usamos a sintaxe de interpolação de string que nos permite incluir variáveis diretamente em uma string usando chaves {}. Ao colocar o sinal $ antes da string, indicamos ao compilador que ela deve ser interpolação de string.


#### FORMAT STRING
Ainda também igual ao Python, podemos usar o format proximo do C++, colocando entre chaves a posição da variável (e formatações adicionais) e, findada a string, após a virgula, as variáveis inseridas no posicionamento.

```CS
Console.WriteLine("Olá, {0}!", nome);
```

> Neste caso, usamos um formato de string que nos permite substituir marcadores de posição {0}, {1}, etc. por valores específicos. Ao passar a variável nome como segundo argumento para o método Console.WriteLine, ela será substituída no marcador de posição {0}.

#### Colection: List
- Sintaxe:
  - Lista Vazia: `List<type> var = new List<type>();`
    - Note que aqui estamos instanciando uma nova lista do tipo determninado para a variável que contenha o mesmo tipo (`List<type>`), sendo o `type` um tipo de dado.
    - Lembrando que `List<type>` também é um tipo, podendo existir uma Lista de Listas.
  - Lista Fixa: `List<type> var = [elemento1, elemento2, ...];`
  - Lista Flexível com valores pré-definidos: `List<type> var = new List<type> { elemento1, elemento2, ... };` 

Ou seja, diremos que uma variável é do tipo Lista de Tipos (string, int, etc).

- Operação:
  - Neste módulo, apresentou-se o método `.Add`, que aplicada a uma lista nós estaremos adicionadno um elemento.

#### Boas práticas
- Nome de Variaveis: Lembrando que é importantíssimo definirmos nomes significativos, ou seja:
  - Nome não significativo: `List<string> string = new List<string>()`;
    - Estamos dizendo que é uma lista de strings com string, mas, não sabemos do que exatamente.
  - Nome significativo: `List<string> listaBandas = List<string>()`;
    - Já aqui sabemos que é uma Lista de Strings que trata-se de uma lista de bandas.

### FOREACH PARA CADA MÚSICA

- Foreach: Para cada elemento da lista;

Resolvendo com For:
```CS
// listaBandas.Count retorna a quantidade de elementos na lista
// Ou seja, retorna a contagem dos elementos internos
for (int i = 0; i < listaBandas.Count; i++)
{
    Console.WriteLine($"Banda: {listaBandas[i]}");
}
```

Resolvendo ForEach:
```CS
foreach (string banda in listaBandas)
{
    Console.WriteLine($"Banda: {banda}");
}
```
- Pontos positivos:
  - Leitura mais clara, mais próxima da humana, frente ao For;
  - Não precisamos passar quantidade da lista, já que foreach percorrerá todos os elementos dela, do inico ao fim, um por um.

- Considerações:
  - Acessa os elementos através da variavel declarada antes da palavra reservada `in`, ou seja, o elemento atual iterado será chamado com o nome declarado
    - Exemplo: string elemento in listaElemento, Console.WriteLine(elemento), ou seja, iterará por cada elemento onde o item atual iterado se chama "elemento". Primeira iteração: Chama o primeiro item (elemento), segunda iteração: Chama o segundo item (elemento) e por ai vai.
  - Recomendado para quando queremos percorrer uma lista de inicio ao fim, aplicando algum comando sobre cada elemento da lista.
  - Não recomendado para quando necessita utilizar os indices da lista. Neste caso, o For é mais recomendado.

#### Qual usar? for ou foreach? (Alura)
Geralmente, é recomendado utilizar o foreach quando se deseja apenas iterar sobre os elementos de uma coleção, sem a necessidade de acessar um índice numérico específico. Isso torna o código mais legível e menos propenso a erros, além de ser mais simples de implementar.

Já o for é recomendado quando é necessário acessar elementos em posições específicas da coleção ou quando se deseja iterar utilizando um índice numérico específico.

##### Refatorando a exibição dos títulos
Quando estamos repetindo ações e tendo que realizar ajustes manuais, este é um bom indicativo que ali necessita-se de uma refatoração para garantir escalabilidade.
- No caso, aconteceu na disposição dos asteríscos
- Ter essa reflexão é importante para garantir menos ajustes manuais quando novos elementos entrarem na aplicação. Ou seja, caso necessite de alguma manutenção, será em apenas um local seja para 2 elementos ou 50 elementos.

Antes de refatorar:
```CS
Console.WriteLine("**************************************");
Console.WriteLine("Exibindo todas as bandas registradas");
Console.WriteLine("**************************************\n");
```

Depois:
```CS
ExibirTituloOpcao("Exibindo todas as bandas registradas");

void ExibirTituloOpcao(string titulo)
{
    int quantidadeDeLetras = titulo.Length;
    string asteriscos = string.Empty.PadLeft(quantidadeDeLetras, '*');
    Console.WriteLine(asteriscos);
    Console.WriteLine(titulo);
    Console.WriteLine(asteriscos + "\n");
}
```

Agora, independente do titulo, os asteriscos em cima e embaixo serão preenchidos com o tamanho da string (quantidade de letras).
- `string.Empty.PadLeft(qtd_de_elementos, elemento_preenchimento)`: string.Empty criamos uma string vazia e temos também o método PadLeft (PadRight e afins), o qual preenche na direita ou na esquerda, com o elemento desejado, a quantidade de elementos. Então, se fosse (3, '+'), o retorno seria "+++". Com a variável, a tornamos dinâmica. **Note que utilizamos, neste caso, aspas simples pra indicar que é um literal. Aparentemente é necessário utilizar aspas simples quando desejamos indicar apenas um elemento (caracter ao invés de string)**.

### CRIANDO O DICIONÁRIO

Para colocarmos as notas em cada banda trabalharemos com a estrutura de dados `Dicionário`. Sua estrutura no C# é similar ao do Python, sendo uma estrutura de chave:valor.

Segue o comando para nós criarmos um dicionario onde a chave são strings e seus valores são uma lista de inteiros. Por ser fortemente tipado, essa declaração é necessária.
```CS
// Dictionary<Tkey, Tvalue>, ou seja, Tipo da chave e Tipo do valor 
Dictionary<string, List<int>> bandasRegistradas = new Dictionary<string, List<int>>();
```

### AJUSTANDO PARA O DICIONÁRIO

```CS
// ANTES
void RegistrarBandas()
{
    listaDeBandas.Add(nomeBanda);
}

// DEPOIS
void RegistrarBandas()
{
    bandasRegistradas.Add(nomeBanda, new List<int>());
}
```

Dicionário, assim como Lista, também possui método Add para adicionar em seu elemento, porém, com 2 parâmetros. Como estamos na função `RegistrarBandas()`, não é de sua responsabilidade atribuir nota, logo, apenas registraremos o nome da banda com uma lista vazia.

Refatorando o foreach
```CS
// ANTES
foreach (string banda in listaDeBandas)
{
    Console.WriteLine($"Banda: {banda}");
}

// DEPOIS
foreach (string banda in bandasRegistradas.Keys)
{
    Console.WriteLine($"Banda: {banda}");
}
```

Precisamos acrescentar o atributo .Keys ao lado do dicionário, pois isso retornará uma lista com todas as suas chaves (que no caso seria o nome das bandas). Caso contrário, Dicionário é um objeto iterável e dará erro.
