Código da Alura:

```CS
// Inicialmente, vamos definir um número secreto não tão secreto assim, apenas para testar nossa aplicação:
// int numeroSecreto = 42;

// Agora, não queremos manter o 42 como número secreto sempre, então vamos criar uma forma de gerar um número pseudo-aleatório:
Random aleatorio = new Random();
int numeroSecreto = aleatorio.Next(1, 101);

// Agora, desenvolvendo a lógica do jogo, podemos usar o seguinte código:
do {
    Console.Write("Digite um número entre 1 e 100: ");
    int chute = int.Parse(Console.ReadLine()!);

    if (chute == numeroSecreto)
    {
        Console.WriteLine("Parabéns! Você acertou o número.");
        break;
    }
    else if (chute < numeroSecreto)
    {
        Console.WriteLine("O número é maior.");
    }
    else
    {
        Console.WriteLine("O número é menor");
    }
} while(true);

// Caso o chute seja igual ao número secreto, vamos exibir a mensagem indicando que o jogo acabou:
Console.WriteLine("O jogo acabou. Você acertou o número secreto!");


```

1. Criar um dicionário que represente um aluno, com uma lista de notas, e mostre a média de suas notas na tela.
```CS
// MINHA RESOLUÇÃO
Dictionary<string, List<float>> alunos = new Dictionary<string, List<float>>
{
    { "Fulano", new List<float>() { 8, 7, 6 } },
    { "Ciclano", new List<float>() { 7, 6, 5 } },
    { "Beltrano", new List<float>() { 9, 8, 8 } }
};


foreach (string pessoa in alunos.Keys)
{
    Console.Write($"{pessoa} tem as seguintes notas:");
    List<float> listaNotas = alunos[pessoa];

    foreach (float nota in listaNotas)
    {
        Console.Write($" {nota}");
    }
    Console.Write($" | Sua média é: {listaNotas.Average():N2}");
    Console.WriteLine();
}

```

3. Crie um programa que implemente um quiz simples de perguntas e respostas. Utilize um dicionário para armazenar as perguntas e as respostas corretas.
```CS
// RESOLUÇÃO ALURA
Dictionary<string, string> perguntasERespostas = new Dictionary<string, string>
{
    { "Qual é a capital do Brasil?", "Brasília" },
    { "Quanto é 7 vezes 8?", "56" },
    { "Quem escreveu 'Romeu e Julieta'?", "William Shakespeare" },
    // Adicione mais perguntas e respostas conforme necessário
};

int pontuacao = 0;

foreach (var pergunta in perguntasERespostas)
{
    Console.WriteLine(pergunta.Key);
    Console.Write("Sua resposta: ");
    string respostaUsuario = Console.ReadLine();

    if (respostaUsuario.ToLower() == pergunta.Value.ToLower())
    {
        Console.WriteLine("Correto!\n");
        pontuacao++;
    }
    else
    {
        Console.WriteLine($"Incorreto. A resposta correta é: {pergunta.Value}\n");
    }
}

Console.WriteLine($"Pontuação final: {pontuacao} de {perguntasERespostas.Count}");
```

Para manipular elementos na lista em C#, é necessário acessar os elementos usando índices ou chaves, por exemplo. É possível adicionar elementos usando o método Add(), remover elementos usando o método Remove() ou alterar elementos diretamente através do seu índice ou chave. Confira abaixo:

```CS
Dictionary<string, List<int>> vendasCarros = new Dictionary<string, List<int>> {
    { "Bugatti Veyron", new List<int> { 10, 15, 12, 8, 5 } },
    { "Koenigsegg Agera RS", new List<int> { 2, 3, 5, 6, 7 } },
    { "Lamborghini Aventador", new List<int> { 20, 18, 22, 24, 16 } },
    { "Pagani Huayra", new List<int> { 4, 5, 6, 5, 4 } },
    { "Ferrari LaFerrari", new List<int> { 7, 6, 5, 8, 10 } }
};
```

Com base na lista de super carros fornecida acima, onde a chave é o nome do carro e os valores são quantas vezes esse carro foi vendido por ano, calcule a média de uma chave específica.

Média de uma chave específica
Para calcular a média de uma chave específica, podemos utilizar o método Average da classe List do C#, conforme o exemplo abaixo, onde calculamos a média de vendas do Bugatti Veyron, por exemplo:
```CS
// RESOLUÇÃO ALURA
double mediaVendasBugatti = vendasCarros["Bugatti Veyron"].Average();
Console.WriteLine("Média de vendas do Bugatti Veyron: " + mediaVendasBugatti);
```
