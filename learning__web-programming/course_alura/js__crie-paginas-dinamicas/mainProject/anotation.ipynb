#### Diferen√ßa de const, var e let (by Copilot)

Vamos explorar as diferen√ßas entre `const`, `var` e `let` no contexto do JavaScript:

1. **`var`**:
   - **Escopo**: A palavra-chave `var` possui **escopo global** ou de fun√ß√£o. Isso significa que a vari√°vel declarada com `var` √© acess√≠vel em todo o escopo da fun√ß√£o onde foi criada.
   - **Hoisting**: Vari√°veis declaradas com `var` s√£o **hoisted** (elevadas) at√© o topo do seu contexto de execu√ß√£o. Isso permite que voc√™ use a vari√°vel antes mesmo de declar√°-la explicitamente.
   - **Inicializa√ß√£o**: A vari√°vel declarada com `var` √© automaticamente inicializada com o valor `undefined` (caso nenhum outro valor seja atribu√≠do).

2. **`let`**:
   - **Escopo**: A palavra-chave `let` tem **escopo de bloco**. Isso significa que a vari√°vel declarada com `let` √© acess√≠vel apenas dentro do bloco onde foi criada (por exemplo, dentro de um loop `for`).
   - **Hoisting**: Assim como `var`, as vari√°veis declaradas com `let` tamb√©m s√£o hoisted, mas n√£o podem ser acessadas antes de sua inicializa√ß√£o.
   - **Inicializa√ß√£o**: A vari√°vel declarada com `let` n√£o √© automaticamente inicializada. Voc√™ precisa atribuir um valor explicitamente.

3. **`const`**:
   - **Escopo**: A palavra-chave `const` declara uma **vari√°vel constante**. Ela tamb√©m tem escopo de bloco, assim como `let`.
   - **Mutabilidade**: Vari√°veis declaradas com `const` n√£o podem ser reatribu√≠das ap√≥s a inicializa√ß√£o. No entanto, se a vari√°vel for um objeto ou array, seus elementos podem ser modificados.

Em resumo:
- Use `const` para valores que n√£o devem ser alterados ap√≥s a inicializa√ß√£o.
- Use `let` quando precisar reatribuir valores.
- Evite usar `var` devido ao seu escopo global e comportamento de hoisting, a menos que haja uma raz√£o espec√≠fica para us√°-lo.

Al√©m dessas tr√™s palavras-chave, existem outras formas de atribui√ß√£o e manipula√ß√£o de vari√°veis no JavaScript, mas essas s√£o as principais. Se tiver mais alguma d√∫vida, estou √† disposi√ß√£o! üòä

## <span style="color: #87BBA2">04. Iterando em listas</span>

Utilizaremos a estrutura `while` para percorrer a listaDeTeclas e realizar as atribui√ß√µes automatizadas e de possibilitando a escalabilidade

````python
// Pom ------------------------------------------------
function tocaSomPom(){
    // Bloco ou corpo da fun√ß√£o
    document.querySelector('#som_tecla_pom').play();
}

// Declara√ß√£o de refer√™ncia
const listaDeTeclas = document.querySelectorAll('.tecla'); 

// USANDO LA√áO DE REPETI√á√ÇO WHILE
// while espera uma condi√ß√£o booleana, ou seja, que retorna se √©
// verdadeira (true) ou falsa (false).

// Para a nossa valida√ß√£o, precisamos criar uma l√≥gica de contador
// comparando-o com o tamanho total de nossa lista. Ou seja,
// enquanto o contador for menor que o tamanho da lista,
// realize o c√≥digo no corpo do while.

let i = 0; // Usaremos o indice como contador

// Note que estamos fazendo compara√ß√£o entre duas refer√™ncias
// sendo o indice (contador) e tamanho da lista.
// Note que usamos .length para acessar a propriedade de
// tamanho da array.
while(i < listaDeTeclas.length) {
    listaDeTeclas[i].onclick = tocaSomPom; // Passamos o "i" como
            // indice iter√°vel

    i++; // mesmo que i = i + 1;
         // Ou seja: valorNovo = valorAntigo + 1;

    console.log(i); // imprime valor no console do devtools
                    // Faremos isso para podermos acompanhar
                    // sua opera√ß√£o
}
````

    - Declara√ß√£o de vari√°vel incremental (contador);
