## <span style="color: #87BBA2">02. Commit, VSCode e equipe</span>

### VSCode no GitHub
Na pagina inicial dentro de um repositório de sua autoria (ou que você tenha permissão), aperte a tecla `.`. Este é o atalho para abrir o **VSCode Web (VSCode na Núvem)** com a estrutura do projeto no commit mais recente da branch que que você está.

Todas as alterações que estão feitas no **VSCode Web** estão sendo "escutadas" pelo GitHub, aparecendo, no simbolo das "branchs" no menu lateral um número indicando quantas alterações no projeto aconteceram (seja criação ou edição de código). Porém, as alterações estão sendo rastreadas mas só subirão para o repositório quando realizarmos o **commit**.

Para commitar, vá no simbolo das branchs, escreva a mensagem de commit na caixinha e **clique no simbolo de correto** (Mas, agora tem um botãozão também).

### Git status, commit e push

Importante ter o código que está no GitHub, o que você está trabalhando, também em sua máquina, pois, como o Git é um sistema distribuido, ele garantirá que o código que está no repositório remoto seja o mesmo que esteja em sua maquina local.

Por isso é importante saber comandos de linha.

Neste momento, quando se está trabalhando localmente, que conheceremos o `push`, que é o **empurrar o que se está localmente para o remoto**. Quando trabalhamos diretamente na web, não tem essa necessidade pois seria meio que trabalhar na raiz do projeto.

#### git pull
Diferente do `git clone`, que apaga TUDO o que você tem e deposita o repositório onde você clonou, o `git pull`.

O `git clone` só se faz para no inicio do projeto, quando se está montando o ambiente local.

```JAVASCRIPT
// Na pasta local administrada por git, no terminal
git pull <link>
```

Retornos:
- Already up to date: Código local em dia com o código no repositório remoto.

#### git status
Verifica o estado atual do código local: Se há alterações feitas e quais foram essas alterações (criação, modificação, deleção).

#### git commit
Comando em terminal, dentro de um ambiente local controlado por git, para realizar o commit das informações previamente adicionadas ou adicionar e commitar juntamente.
```JAVASCRIPT
// Adicionando e commitando juntos
git commit <arquivo_modificado_ou_criado> -m "<mensagem>"
// Pra adicionar e commitar todas as alterações
git commit . -m "<mesangem>"
```

#### git push
Seguindo todos os passos anteriores, atualizamos nosso projeto localmente mas não remotamente. Para "empurrarmos" até o repositório local nós usamos os `git push`.
```JAVASCRIPT
// Empurrando para a origem
git push origin main
```
Ao realizar esse teste em algum repositório publico, você deve ter enfrentado algum erro. O `git clone` pode ser realizado de qualquer repositório remoto, mas, o `git push` só pode em repositórios onde somos autores ou colaboradores e, além disso, já termos configurado, corretamente, nossos parametros de git em nosso computador.

Basicamente, esses parametros - que ficam em um arquivo de configuração, como sua chave pública e chave primaria - são configurados que se faz uma vez por máquina para gerar sua chave para o gitHub saber quem é a pessoa que está realizando o push, e, com isso, validar se ela pode ou não ter acesso ao repositório o qual busca-se dar o push e realizar seu registro.
> É uma forma de fazermos o handshake e, assim, realizar a operação.
>
> Isso não é realizado no repositório remoto pois no remoto o acesso já está validado, já no local não, aí, isso que estamos fazendo é justamente configurar nossa chave de acesso do repositório local para o remoto.

### O que vimos até aqui

DIFF = A tela de antes e a tela depois de um commit realizado. É o `diff` de um commit, a diferença entre sua versão anterior e a versão atual.

Repare que nosso `git pull` e `git push` não está sendo necessario colocar url, pois já temos nosso repositório local configurado com o remoto e não há mais ramificações além da principal (main).

#### Voltando commits
Para não precisar passar por cima dos commits, ou seja, dando ctrl+c e ctrl+v e dando um novo commit, o Git possibilita retornarmos a commits anteriores.

Para fazer tudo em linha de comandos:
```JAVASCRIPT
git log --oneline

// relatório de hash e comentarios dos commits

// Caso queira retornar a versão de um arquivo específico
git restore --source <hash_do_commit> <nome_arquivo>
```
```JAVASCRIPT
// Caso queira retornar a versão de todos os arquivos
// para uma versão específica
git restore --source <hash_do_commit> .
```

Em seguida, demos o `git push` e não funcionou! Não funcinou pois realizamos as alterações e não as commitamos (não criamos seu rótulo).

Commitamos o restart e demos o `git push` e agora sim a operação foi realizada em nosso repositório remoto.

Precisamos deixar nosso repositório local pronto e limpo para enviar ao remoto, o local **nunca pode estar atrás** do remoto e, sim, **à frente**, pois, quando estamos o local está à frente aí sim podemos dar o commit.


## <span style="color: #87BBA2">04. Adicionando arquivos</span>

### git add
Essa é uma etapa antes do commit, usada para preparar os arquivos para serem commitados. Sem essa adição, o git não rastreia seu versionamento.

Caso tenha alguma alteração, o ideal é adicionar o arquivo com alteração desejado, ou todos os arquivos, e em seguida realizar a mesma operação mas com commit.

Caso queira atualizar aquivo(s) específico(s) com alteração:
```JAVASCRIPT
// Verifica se o repositório local possui alguma mudança frente ao remoto
git status
// Tem alguma alteração que está na frente do remoto? Adicionando arquivos
git add <arquivo_especifico_alterado>
// Comitando o arquivo adicionado
git commit <arquivo_especifico_adicionado> -m "<mensagem>"
// Empurrando para o repositório remoto
git push
```

Caso queira atualizar todos os aquivos específicos com alteração:
```JAVASCRIPT
// Verifica se o repositório local possui alguma mudança frente ao remoto
git status
// Tem alguma alteração que está na frente do remoto? Adicionando arquivos
git add .
// Comitando o arquivo adicionado
git commit -m "<mensagem>"
// Empurrando para o repositório remoto
git push
```
